### Import libraries

```python
import pandas as pd 
import numpy as np 
import yfinance as yf 

import matplotlib as mlp
import matplotlib.pyplot as plt 
%matplotlib inline
plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings('ignore')
```

### Load stock data

```python
# stock list

stock_list = ['AAPL', 'NVDA', 'MSFT', 'WDC']
len(stock_list)
```

```python
type(stock_list)
```

```python
pd.Series(stock_list)
```

```python
# setting
start = '2000-01-01'
end = '2023-12-31'
interval = '1wk' # get data by weekly

np.random.seed(42)
N = 3

sample_stock_list = pd.Series(stock_list).sample(N).to_list()
sample_stock_list 
```

```python
# fetch data

df = yf.download(sample_stock_list, start = start, interval = interval, end = end)
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df = df['Adj Close']
df.head(10)
```

```python
# plot

mlp.rcParams['figure.figsize'] = (10, 5)
df.plot()
plt.title(f'Sample stock from {start} to {end}')
plt.xlabel('Price')
plt.ylabel('Date')
plt.show()
```

```python
# calculate simple return

for ticker in sample_stock_list:
    df[f'{ticker}_returns'] = df[f'{ticker}'].pct_change()

df
```

```python
# setting

n_lookback = 6
n_sample = len(sample_stock_list)
```

```python
n_sample
```

```python
my_stock_list = {
    f"{sample_stock_list[0]}_close":df[sample_stock_list[0]].copy(),
    f"{n_lookback}_weeks_ago":df[sample_stock_list[0]].shift(n_lookback)
}
my_stock_list
```

```python
len(my_stock_list)
```

```python
# store my stock list into a dataframe

sample_df = pd.DataFrame(data = my_stock_list)
sample_df
```

```python
# add signal column to a dataframe

sample_df['single'] = np.where(sample_df[f'{sample_stock_list[0]}_close'] > sample_df[f'{n_lookback}_weeks_ago'], 1, 0)
sample_df 
```

```python
sample_df.head(20)
```

```python
# select data from look back date to another 15 days

sample_df.iloc[n_lookback : n_lookback + 15]
```
