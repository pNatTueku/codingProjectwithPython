```python
# setting
start = '2000-01-01'
end = '2023-12-31'
interval = '1wk' # get data by weekly

np.random.seed(42)
N = 3

sample_stock_list = pd.Series(stock_list).sample(N).to_list()
sample_stock_list 
```

```python
# fetch data

df = yf.download(sample_stock_list, start = start, interval = interval, end = end)
df.head(10)
```

```python
df = df['Adj Close']
df.head(10)
```

```python
# plot

mlp.rcParams['figure.figsize'] = (10, 5)
df.plot()
plt.title(f'Sample stock from {start} to {end}')
plt.xlabel('Price')
plt.ylabel('Date')
plt.show()
```

```python
# calculate simple return

for ticker in sample_stock_list:
    df[f'{ticker}_returns'] = df[f'{ticker}'].pct_change()

df
```

```python
# setting

n_lookback = 6
n_sample = len(sample_stock_list)
```

```python
n_sample
```

```python
my_stock_list = {
    f"{sample_stock_list[0]}_close":df[sample_stock_list[0]].copy(),
    f"{n_lookback}_weeks_ago":df[sample_stock_list[0]].shift(n_lookback)
}
my_stock_list
```

```python
len(my_stock_list)
```

```python
# store my stock list into a dataframe

sample_df = pd.DataFrame(data = my_stock_list)
sample_df
```

```python
# add signal column to a dataframe

sample_df['single'] = np.where(sample_df[f'{sample_stock_list[0]}_close'] > sample_df[f'{n_lookback}_weeks_ago'], 1, 0)
sample_df 
```

```python
for i in range(n_lookback, len(df)-1):
    
    # print(f"for week: {i}, \nindex: {df.index[i]}")
    weights  = np.array([0.0] * n_sample)
    for j, ticker in enumerate(sample_stock_list): # j for index, ticker for data
        # print(ticker)
        # print(j)

        if df[ticker][i] > df[ticker][1 - n_lookback]: # check by weekly
            weights[j] = 1 # j is for each ticker, 1 is a signal to buy
        else:
            weights[j] = 0

    if sum(weights) != 0: # buy signal
        weights /= sum(weights) # devide
    print(weights)
```

```python
for i in range(n_lookback, len(df)-1):
    
    print(f"for week: {i}, \nindex: {df.index[i]}")
    weights  = np.array([0.0] * n_sample)
    for j, ticker in enumerate(sample_stock_list): # j for index, ticker for data
        print(ticker)
        # print(j)

        if df[ticker][i] > df[ticker][1 - n_lookback]: # check by weekly
            weights[j] = 1 # j is for each ticker, 1 is a signal to buy
        else:
            weights[j] = 0

    if sum(weights) != 0: # buy signal
        weights /= sum(weights) # devide
    print(weights)

    # here we check the profit
    equity = 0 # set an initial value
    print(f"returns in: {df.index[i+1]}") 
    
    for j, ticker in enumerate(sample_stock_list):
        
        print(f'{ticker} returns: {df[f"{ticker}_returns"][i+1]}')
        equity += df[f"{ticker}_returns"][i+1] * weights[j] # multiply by weight
```
