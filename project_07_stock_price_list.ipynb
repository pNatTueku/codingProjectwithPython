### Import libraries

```python
import pandas as pd 
import numpy as np 
import yfinance as yf 

import matplotlib as mlp
import matplotlib.pyplot as plt 
%matplotlib inline
plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings('ignore')
```

### Load stock data

```python
# stock list

stock_list = ['AAPL', 'NVDA', 'MSFT', 'WDC']
len(stock_list)
```

```python
type(stock_list)
```

```python
pd.Series(stock_list)
```

```python
type(pd.Series(stock_list))
```

```python
# setting
start = '2000-01-01'
end = '2023-12-31'
interval = '1wk' # get data by weekly

np.random.seed(42)
N = 3

sample_stock_list = pd.Series(stock_list).sample(N).to_list()
sample_stock_list 
```

```python
# fetch data

df = yf.download(sample_stock_list, start = start, interval = interval, end = end)
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df = df['Adj Close']
df.head(10)
```

```python
# plot

mlp.rcParams['figure.figsize'] = (8, 5)
df.plot()
plt.title(f'Sample stock from {start} to {end}')
plt.xlabel('Price')
plt.ylabel('Date')
plt.show()
```

```python
# calculate simple return

for ticker in sample_stock_list:
    df[f'{ticker}_returns'] = df[f'{ticker}'].pct_change()

df
```
