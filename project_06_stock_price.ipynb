### Import libraries

```python
# import library

import yfinance as yf

import matplotlib as mlp
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

### Fetching data

```python
# setting

ticker = "NVDA"
start = '2000-01-01'
end = '2023-12-31'
```

```python
# load data

df = yf.download(tickers = ticker, start = start, end = end)
df.head(10)
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# plot

mlp.rcParams['figure.figsize'] = (8, 5)
df['Close'].plot()
plt.title(f'{ticker} from {start} to {end}')
plt.show()
```

### Strategy

```python
# setting

look_back = 100
```

```python
# get the highest close price referred to the look back number

df['highest'] = df['Close'].rolling(look_back).max()
df.head()
```

```python
df.tail()
```

```python
df.head(100)
```

```python
df.head(101)
```

```python
# plot

mlp.rcParams['figure.figsize'] = (8, 5)
df[['Close', 'highest']].plot()
plt.title(f'{ticker} from {start} to {end}')
plt.legend()
plt.show()
```
