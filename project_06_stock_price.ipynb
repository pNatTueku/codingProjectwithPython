### Import libraries

```python
# import library

import yfinance as yf

import matplotlib as mlp
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

### Fetching data

```python
# setting

ticker = "NVDA"
start = '2000-01-01'
end = '2023-12-31'
```

```python
# load data

df = yf.download(tickers = ticker, start = start, end = end)
df.head(10)
```

### Strategy

```python
# setting

look_back = 100
```

```python
# get the highest close price referred to the look back number

df['highest'] = df['Close'].rolling(look_back).max()
df.head()
```

```python
# plot

mlp.rcParams['figure.figsize'] = (8, 5)
df[['Close', 'highest']].plot()
plt.title(f'{ticker} from {start} to {end}')
plt.legend()
plt.show()
```

```python
df.iloc[90:120]
```

```python
# shift row data

df['highest_shift'] = df['highest'].shift(1)
df.iloc[90:110]
```

```python
# select 2 collumns to plot

mlp.rcParams['figure.figsize'] = (8, 5)
df[['Close', 'highest_shift']].loc["2022-04": "2023"].plot()
plt.title(f'{ticker} from {start} to {end}')
plt.legend()
plt.show()
```

```python
# stop percentage 90% - we will sale after 10% of price dropping
stop_pct = 0.9

mlp.rcParams['figure.figsize'] = (8, 5)
df[['Close', 'highest_shift']].loc["2022-04": "2023"].plot()
(df['highest_shift'] * stop_pct).loc["2022-04": "2023"].plot(label = f'stop {(1 - stop_pct) * 100:.2f}%')
plt.title(f'{ticker} from {start} to {end}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# add signal column

df['signal'] = 0
df.head()
```

```python
df.tail()
```

```python
for i in range(look_back + 1, len(df)): # look_back + 1 (select 2 days ago High)
    
    # Price will be equal to today price if yesterday price equal or less than today price
    if (df["Close"].iloc[i] > df["highest_shift"].iloc[i]) & (df["Close"].iloc[i-1] <= df["highest_shift"].iloc[i-1]): 
        df["signal"].iloc[i] = 1 # if > previous High price then buy by checking yesterday close price if it less or equal to then we know that it's breaking
    # If today price is less than stop loss 10% then 0
    elif (df["Close"].iloc[i] < df["highest_shift"].iloc[i] * stop_pct):
        df["signal"].iloc[i] = 0 
    # If price not in 2 above conditions, lets maintain signal
    else:
         df["signal"].iloc[i] =  df["signal"].iloc[i-1]        
            
df.head()
```
